```python
x=df[['engine-size','horsepower','width']].values
x
```

```python
y=df['price'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=0) # split data into training and testing
```

```python
from sklearn.linear_model import LogisticRegression
linreg = LogisticRegression()
linreg.fit(xtrain, ytrain)
ypred= linreg.predict(xtest)
linreg.score(xtrain, ytrain) #training accuracy
```

Accuracy of model is 0.37

Evaluating model accurarcy

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score
```

```python
print(f"MAE -: {maetest, ypred)}")
print(f"MSE -: {mse(ytest, ypred)}")
print(f"RMSE -: {np.sqrt(mse(ytest, ypred))}")
print(f"R-squared -: {r2_score(ytest, ypred)}") #test accuracy(y
```

Hyperparameter tuning

```python
n = list(range(1,31))

accuracy = []


for i in n:
    linreg=LogisticRegression(C=i,max_iter=1000)
   #linreg = LinearRegression(n_jobs=i)
    linreg.fit(xtrain, ytrain)
    ypred = linreg.predict(xtest)
    a =r2_score(ytest,ypred) 
    accuracy.append(a)
```

```python
plt.plot(n,accuracy)
plt.xlabel('Value of n')
plt.ylabel('Accuracy level')
plt.title('Accuracy w.r.t. n')
plt.grid(True)
plt.show()
```
